```python
#creating a function
def clean_review(ini_review):
    #remove html
    no_html = BeautifulSoup(ini_review).get_text()
    
    #keep words only
    words = no_html.split()
    table = str.maketrans('', '', string.punctuation)
    removed_punc = [w.translate(table) for w in words]
    join_review = ' '.join(removed_punc)
    
    #noramlize to lowercase
    words_lower = join_review.lower().split()
    
    #removing stop words
    words_lower = [i for i in words_lower if not i in stopwords.words("english") ]
    
    return (" ".join(words_lower))
```

```python
print(clean_review(reviews_data["review"][15]))
```

```python
print(clean_review(reviews_data["review"][0]))
```

```python
#using the function to clean all the reviews in the data
clean_data = []
num = reviews_data["review"].size
```

```python
for i in range(0,num):                                                                            
    clean_data.append( clean_review( reviews_data["review"][i] ))
```
